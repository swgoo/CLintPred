```python
import os
import pandas as pd

def convert_supplemental_trainset(df_load):
    df_trainset = df_load.drop(df_load[df_load['human CL (mL/min/kg)'].notna()].index)

    df_trainset.drop("human CL (mL/min/kg)", axis=1, inplace=True)
    df_trainset.rename(columns = {'human CL (mL/min/kg)':'Clint'},inplace=True)
    return df_trainset

def convert_supplemental_testset(df_load):
    df_testset = df_load.drop("human CL (mL/min/kg)", axis=1)
    
    df_testset.rename(columns = {'human CL (mL/min/kg)':'Clint'},inplace=True)
    df_testset.dropna(subset=['Clint'], axis=0, inplace=True)
    return df_testset
```
